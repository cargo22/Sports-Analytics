Now, obviously it's impossible to tell just how good this kid can become. However, a criticism I often see on social media is Yamal's finishing ability. Most critiques I see around Yamal go along the lines of "if he was better at finishing..." Now, as a diehard Barcelona fan, I watch almost every game. I too notice that Yamal's finishing seems to be the area he lacks the most. I've seen him miss plenty of open nets and solo 1v1s with the keeper. However, is his finishing really that bad?

I will dive deeper into what separates the best finishers in world football, and see what conclusion we can derive. Is Yamal truly that poor of a finisher?

To carry out this short research question, I will gather the best finishers in world football this season (aka the top 10 league scorers this season), and see what their goal count is in comparison to their xG. In other words, the question at hand is, **do good goal scorers outperform their xG, and if so, where does Lamine Yamal rank amongst the best?**

```python
# setting up environment to carry out analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# importing the dataframe created
# stats are from StatsMuse https://www.statmuse.com/fc/ask?q=goal+leaders+in+europe+top+five+2024-25
df = pd.read_csv('xG tracker - Sheet1.csv',header=None)

# time to do a little data cleaning
df.columns = ['Player', 'Goals Scored', 'xG']

df.head()
```

Our Database includes the top 20 league scorers of this season (2024-2025), accompanied with the amount of goals they scored and their xG.

```python
# making sure the values are actually numeric
df['Goals Scored'] = pd.to_numeric(df['Goals Scored'])
df['xG'] = pd.to_numeric(df['xG'])
```

```python
sns.regplot(data=df, x='xG', y='Goals Scored')
plt.title("Goals Scored vs xG | Top 20 Goal Scorers 24-25")
plt.xlabel("xG")
plt.ylabel("Goals Scored")
```
