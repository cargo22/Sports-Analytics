To carry out this short research question, I will gather the best finishers in world football this season (aka the top 20 league scorers this season), and see what their goal count is in comparison to their xG. In other words, the question at hand is, **do good goal scorers outperform their xG, and if so, where does Lamine Yamal rank amongst the best?**

```python
# setting up environment to carry out analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# importing the dataframe created
# stats are from StatsMuse https://www.statmuse.com/fc/ask?q=goal+leaders+in+europe+top+five+2024-25
df = pd.read_csv('xG tracker - Sheet1.csv',header=None)

# time to do a little data cleaning
df.columns = ['Player', 'Goals Scored', 'xG']

df.head()
```

Our Database includes the top 20 league scorers of this season (2024-2025), accompanied with the amount of goals they scored and their xG.

```python
# making sure the values are actually numeric
df['Goals Scored'] = pd.to_numeric(df['Goals Scored'])
df['xG'] = pd.to_numeric(df['xG'])
```

Now, we will keep track of the ratio of [Goals Scored: xG]. That is, if a player's raio is above 1, they are considered a good finisher, as they score more than what they are expected to.

```python
# extracting the goals scored to expected goals ratio
df['xG ratio'] = df['Goals Scored'] / df['xG']

# sorting to see who exceeded their expected goals the most
df = df.sort_values(by = 'xG ratio', ascending = False)

df.head()
```

```python
# plotting ratios by players
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Player', y='xG ratio')

plt.xticks(rotation=45, ha='right')
plt.title('Goal Conversion Efficiency - 2024/25')
plt.ylabel('xG Ratio')
plt.xlabel('Player')
plt.tight_layout()
plt.show()
```

Now, let's insert Lamine Yamal and see where he fits amongst Europe's best scorers this season.

```python
# Lamine's stats are extracted from StatMuse, like all other players
lamine = {
    'Player': 'L. Yamal',
    'Goals Scored': 9,
    'xG': 10.26
}

# calculating Yamal's ratio
lamine['xG ratio'] = lamine['Goals Scored'] / lamine['xG']

# inserting into df and resorting
df.loc[len(df)] = lamine

df.sort_values(by='xG ratio', ascending=False)

df.head()
```

Now that Lamine Yamal has been added to the table, let's see where he lies amongst the top 20 scorers this season.

```python
# plotting ratios by player w/ Yamal
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Player', y='xG ratio')

plt.xticks(rotation=45, ha='right')
plt.title('Goal Conversion Efficiency - 2024/25')
plt.ylabel('xG Ratio')
plt.xlabel('Player')
plt.tight_layout()
plt.show()
```
